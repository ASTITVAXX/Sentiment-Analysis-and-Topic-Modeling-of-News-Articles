```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

# Ignore warnings
warnings.filterwarnings("ignore")

data=pd.read_csv('sentiment_topic_scores.csv')    # compiled1.csv is the file which contains the topics and their score along with sentiment scores
```

```python
data['month'] = data['month'].astype(int)
data['year'] = data['year'].astype(int)
data['joined'] = data['month'].astype(str) + '/' + data['year'].astype(str)
```

```python
topic_number= '2'    
probability_threshold=0 
```

```python
def plot_sentiment_over_time(data, topic_number, probability_threshold):
    topic_column = f'{topic_number}'
    topic = data[data[topic_column] > probability_threshold]
    topic['quarter'] = pd.PeriodIndex(pd.to_datetime(topic['joined'], format='%m/%Y'), freq='Q')
   
    sum = topic[topic_column].sum()
    topic['positive_score'] = topic['positive_score'] * topic[topic_column]/sum
    topic['negative_score'] = topic['negative_score'] * topic[topic_column]/sum

    topic['positive_score'] = topic.groupby(['quarter'])['positive_score'].transform('median')
    topic['negative_score'] = topic.groupby(['quarter'])['negative_score'].transform('median')
    topic = topic.drop_duplicates(subset=['quarter'])
    topic = topic.sort_values(by="quarter")
    topic = topic[(topic["year"] > 2019) | ((topic["year"] == 2019) & (topic["month"] >= 8))]    
    plt.rcParams['figure.figsize'] = [20, 10]
    topic['quarter'] = topic['quarter'].astype(str)

    plt.plot(topic['quarter'], topic['positive_score'], label='Positive')
    plt.xticks(topic['quarter'], rotation=45)
    plt.plot(topic['quarter'], topic['negative_score'], label='Negative')
    plt.title(f"Topic {topic_number+1} Sentiment Over Time")
    # plt.title(f"Overall Sentiment Over Time")
    plt.xlabel('Date')
    plt.ylabel('Sentiment Count')
    plt.legend()
    plt.show()
    # print(topic)
```

```python
def plot_topic_usage(data, topic_number, probability_threshold):
    topic_column = f'{topic_number}'
    topic = data[data[topic_column] > probability_threshold]
    topic['quarter'] = pd.PeriodIndex(pd.to_datetime(topic['joined'], format='%m/%Y'), freq='Q')
    topic_usage = topic.groupby('quarter').size().reset_index(name='usage_count')
    # topic_usage = topic_usage[topic_usage['quarter'].dt.year > 2018]
    topic_usage = topic_usage[(topic_usage["quarter"].dt.year > 2019) | ((topic_usage["quarter"].dt.year == 2019) & (topic_usage["quarter"].dt.month >= 8))]    


    plt.rcParams['figure.figsize'] = [10, 6]
    topic_usage['quarter'] = topic_usage['quarter'].astype(str)  # Convert quarter to string
    plt.plot(topic_usage['quarter'], topic_usage['usage_count'])
    plt.title(f"Topic {topic_number+1} Usage Over Time")
    plt.xlabel('Quarter')
    plt.ylabel('Usage Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
k=7 # number of topics
```

```python
for i in range(0, k):   # change the range according to the number of topics
    plot_topic_usage(data, i, 0.2)
```

```python
for i in range(1, k):
    plot_sentiment_over_time(data, i, 0.2)
```
